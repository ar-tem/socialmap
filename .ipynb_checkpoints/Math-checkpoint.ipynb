```python
import matplotlib.pyplot as plt
import random
```

```python
data_x1 = range(1, 100, 5)
data_y1 = range (0, 20)
```

```python
plt.plot(data_x1, data_y1)
plt.show()
```

```python
data_x2 = random.sample(range(500), 10)
data_y2 = random.sample(range(100), 10)
data_x2.sort()
data_y2.sort()
plt.plot(data_x2, data_y2)
plt.show()
```

```python
plt.plot(data_x2, data_y2, color='red', linestyle='dashed')
plt.figure()
plt.plot(data_x1, data_y1, color='green', linestyle='dotted', linewidth=4)
plt.show()
```
